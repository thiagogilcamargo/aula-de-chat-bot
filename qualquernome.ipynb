```python
import cv2 as cv
```

```python
import matplotlib.pyplot as plt
```

```python
path_img = "D:\dados\dados\\"
```

```python
Imagem_BGR = cv.imread(path_img + "lena.png", cv.IMREAD_COLOR)
Imagem_Gray = cv.imread(path_img + "lena.png", cv.IMREAD_GRAYSCALE)
```

```python
thr_otsu = cv.threshold(Imagem_Gray, 0,255,cv.THRESH_BINARY | cv.THRESH_OTSU)[1]
thr_bin = cv.threshold(Imagem_Gray, 150,255,cv.THRESH_BINARY)[1]
```

```python
print("Cor do pix (0,0) da image BGR:", Imagem_BGR[0,0])
print("Cor do pixal (0,0) da image GRayscale:", Imagem_Gray[0,0])
```

```python
import librosa 
import librosa.display
import matplotlib.pyplot as plt

# import IPython
```

```python
path_dataset = "D:\dados\dados\\"
```

```python
sinal, sr=librosa.load(path_dataset + "aeiou_ruido.wav", sr=44100) 
```

```python
librosa.display.waveshow(sinal,sr=sr,color ='r')
plt.grid (color= 'k', linestyle = '-', linewidth=0.3)
plt.xlabel('time (s)', fontsize=14)
plt.ylabel('amplitude (a.u)', fontsize=14)
```
